**Importing Required Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
```

**Loading the Datasets**

```python
# Load the datasets
fertilizer_df = pd.read_csv('Fertilizer Prediction.csv')
crop_recommendation_df = pd.read_csv('Crop_recommendation.csv')

# Display the first few rows of each dataset
print('Fertilizer Prediction Dataset:')
print(fertilizer_df.head())

print('\nCrop Recommendation Dataset:')
print(crop_recommendation_df.head())
```

**General Info about Each Dataset**


```python
# General info about each dataset (e.g., number of rows, columns, types of data)
print('\nFertilizer Dataset Info:')
fertilizer_df.info()

print('\nCrop Recommendation Dataset Info:')
crop_recommendation_df.info()
```

 **Check for Missing Values**


```python
# Check for missing values in each dataset
print('\nMissing values in Fertilizer Dataset:')
print(fertilizer_df.isnull().sum())

print('\nMissing values in Crop Recommendation Dataset:')
print(crop_recommendation_df.isnull().sum())
```

**Summary Statistics**


```python
# Get summary statistics for numerical columns
print('\nSummary Statistics for Fertilizer Dataset:')
print(fertilizer_df.describe())

print('\nSummary Statistics for Crop Recommendation Dataset:')
print(crop_recommendation_df.describe())
```

 **Univariate Analysis: Distribution of Numerical Features in Fertilizer Dataset**


```python
fertilizer_df.hist(bins=15, figsize=(15, 10), color='skyblue')
plt.suptitle('Distribution of Numerical Features in Fertilizer Dataset')
plt.show()
```

**Univariate Analysis: Frequency Count of Categorical Columns in Crop Recommendation Dataset**

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=crop_recommendation_df, x='label')  # Assuming 'label' is the crop type
plt.title('Frequency of Crop Types in Crop Recommendation Dataset')
plt.xticks(rotation=90)
plt.show()
```

**Pairplot for Visualizing Pairwise Relationships in Fertilizer Dataset**

```python
sns.pairplot(fertilizer_df)
plt.suptitle('Pairplot for Fertilizer Dataset', y=1.02)
plt.show()
```

**Pairplot for Visualizing Pairwise Relationships in Fertilizer Dataset**


```python
sns.pairplot(crop_recommendation_df.select_dtypes(include=[float, int]))
plt.suptitle('Pairplot for Crop Recommendation Dataset', y=1.02)
plt.show()
```

 **Boxplot for Detecting Outliers in Fertilizer Dataset**

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=fertilizer_df, orient='h', palette='Set2')
plt.title('Boxplot to Detect Outliers in Fertilizer Dataset')
plt.show()
```

**Boxplot for Detecting Outliers in Crop Recommendation Dataset**


```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=crop_recommendation_df.select_dtypes(include=[float, int]), orient='h', palette='Set2')
plt.title('Boxplot to Detect Outliers in Crop Recommendation Dataset')
plt.show()
```

**Visualizing Missing Data in Fertilizer Dataset**

```python
plt.figure(figsize=(10, 6))
sns.heatmap(fertilizer_df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Data Heatmap for Fertilizer Dataset')
plt.show()
```

**Visualizing Missing Data in Crop Recommendation Dataset**

```python
plt.figure(figsize=(10, 6))
sns.heatmap(crop_recommendation_df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Data Heatmap for Crop Recommendation Dataset')
plt.show()
```

**KDE Plot for Numerical Features of Fertilizer Dataset**

```python
sns.pairplot(fertilizer_df, kind='kde')
plt.suptitle('Pairwise KDE Plots for Fertilizer Dataset', y=1.02)
plt.show()
```

**KDE Plot for Numerical Features of Crop Recommendation Dataset**


```python
sns.pairplot(crop_recommendation_df.select_dtypes(include=[float, int]), kind='kde')
plt.suptitle('Pairwise KDE Plots for Crop Recommendation Dataset', y=1.02)
plt.show()
```

**Detecting Outliers Using Z-Score Method and Removing Them**

```python
# Compute Z-scores for the numerical features
z_scores = zscore(fertilizer_df.select_dtypes(include=[float, int]))

# Identify rows with Z-scores > 3 (common threshold for outliers)
outliers = (z_scores > 3).all(axis=1)

# Removing outliers from the dataset
fertilizer_df_no_outliers = fertilizer_df[~outliers]

# Boxplot to check the data after removing outliers
plt.figure(figsize=(15, 10))
sns.boxplot(data=fertilizer_df_no_outliers, orient='h', palette='Set2')
plt.title('Boxplot After Removing Outliers from Fertilizer Dataset')
plt.show()
```

**Verify Column Type**

```python
# Check column types for the Fertilizer dataset
print("Fertilizer Dataset column types:")
print(fertilizer_df.dtypes)

# Check column types for the Crop Recommendation dataset
print("\nCrop Recommendation Dataset column types:")
print(crop_recommendation_df.dtypes)
```

**Identify Non-Numeric Columns**

```python
# Get non-numeric columns
non_numeric_fertilizer = fertilizer_df.select_dtypes(include=['object']).columns
non_numeric_crop = crop_recommendation_df.select_dtypes(include=['object']).columns

print("\nNon-numeric columns in Fertilizer Dataset:", non_numeric_fertilizer)
print("Non-numeric columns in Crop Recommendation Dataset:", non_numeric_crop)
```

 **Convert Non-Numeric Columns**


```python
from sklearn.preprocessing import LabelEncoder

# Encode non-numeric columns for the Fertilizer dataset
for col in non_numeric_fertilizer:
    fertilizer_df[col] = LabelEncoder().fit_transform(fertilizer_df[col])

# Encode non-numeric columns for the Crop Recommendation dataset
for col in non_numeric_crop:
    crop_recommendation_df[col] = LabelEncoder().fit_transform(crop_recommendation_df[col])
```

**Handle Missing or Invalid Values**

```python
# Check for missing values in both datasets
print("\nMissing values in Fertilizer Dataset:")
print(fertilizer_df.isnull().sum())

print("\nMissing values in Crop Recommendation Dataset:")
print(crop_recommendation_df.isnull().sum())
```

**Verify Dataset for Correlation**

```python
print("Fertilizer Dataset after preprocessing:")
print(fertilizer_df.dtypes)

print("\nCrop Recommendation Dataset after preprocessing:")
print(crop_recommendation_df.dtypes)
```
